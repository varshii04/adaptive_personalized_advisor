```python

arr = []

for index in df2.index:
    if(df2.at[index,"Day"] == "Monday"):
        num = df2.at[index,"Contextual_group"]  + 1
        df1 = pd.read_csv("Dataset/SecondLevel/Cluster " + str(num)  +"/monday.csv", index_col = 0)
        arr.append(df1.at[df2.at[index,"ID"],"Cluster"])
    elif(df2.at[index,"Day"] == "Tuesday"):
        num = df2.at[index,"Contextual_group"]  + 1
        df1 = pd.read_csv("Dataset/SecondLevel/Cluster " + str(num)  +"/tuesday.csv", index_col = 0)
        arr.append(df1.at[df2.at[index,"ID"],"Cluster"])
    elif(df2.at[index,"Day"] == "Wednesday"):
        num = df2.at[index,"Contextual_group"]  + 1
        df1 = pd.read_csv("Dataset/SecondLevel/Cluster " + str(num)  +"/wednesday.csv", index_col = 0)
        arr.append(df1.at[df2.at[index,"ID"],"Cluster"])
    elif(df2.at[index,"Day"] == "Thursday"):
        num = df2.at[index,"Contextual_group"]  + 1
        df1 = pd.read_csv("Dataset/SecondLevel/Cluster " + str(num)  +"/thursday.csv", index_col = 0)
        arr.append(df1.at[df2.at[index,"ID"],"Cluster"])
    elif(df2.at[index,"Day"] == "Friday"):
        num = df2.at[index,"Contextual_group"]  + 1
        df1 = pd.read_csv("Dataset/SecondLevel/Cluster " + str(num)  +"/friday.csv", index_col = 0)
        arr.append(df1.at[df2.at[index,"ID"],"Cluster"])
    elif(df2.at[index,"Day"] == "Saturday"):
        num = df2.at[index,"Contextual_group"]  + 1
        df1 = pd.read_csv("Dataset/SecondLevel/Cluster " + str(num)  +"/saturday.csv", index_col = 0)
        arr.append(df1.at[df2.at[index,"ID"],"Cluster"])
    elif(df2.at[index,"Day"] == "Sunday"):
        num = df2.at[index,"Contextual_group"]  + 1
        df1 = pd.read_csv("Dataset/SecondLevel/Cluster " + str(num)  +"/sunday.csv", index_col = 0)
        arr.append(df1.at[df2.at[index,"ID"],"Cluster"])
    
df2["Activity Pattern"] = arr
```
