```python
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import random


mon = pd.DataFrame(columns=['ID','a0','b0','c0','d0','a1','b1','c1','d1','a2','b2','c2','d2','a3','b3','c3','d3','Average Daily Steps','No of active session','A_Mean','A_Median','A_Mode','No of sedentary session','S_Mean','S_Median','S_Mode'])
tue = pd.DataFrame(columns=['ID','a0','b0','c0','d0','a1','b1','c1','d1','a2','b2','c2','d2','a3','b3','c3','d3','Average Daily Steps','No of active session','A_Mean','A_Median','A_Mode','No of sedentary session','S_Mean','S_Median','S_Mode'])
wed = pd.DataFrame(columns=['ID','a0','b0','c0','d0','a1','b1','c1','d1','a2','b2','c2','d2','a3','b3','c3','d3','Average Daily Steps','No of active session','A_Mean','A_Median','A_Mode','No of sedentary session','S_Mean','S_Median','S_Mode'])
thu = pd.DataFrame(columns=['ID','a0','b0','c0','d0','a1','b1','c1','d1','a2','b2','c2','d2','a3','b3','c3','d3','Average Daily Steps','No of active session','A_Mean','A_Median','A_Mode','No of sedentary session','S_Mean','S_Median','S_Mode'])
fri = pd.DataFrame(columns=['ID','a0','b0','c0','d0','a1','b1','c1','d1','a2','b2','c2','d2','a3','b3','c3','d3','Average Daily Steps','No of active session','A_Mean','A_Median','A_Mode','No of sedentary session','S_Mean','S_Median','S_Mode'])
sat = pd.DataFrame(columns=['ID','a0','b0','c0','d0','a1','b1','c1','d1','a2','b2','c2','d2','a3','b3','c3','d3','Average Daily Steps','No of active session','A_Mean','A_Median','A_Mode','No of sedentary session','S_Mean','S_Median','S_Mode'])
sun = pd.DataFrame(columns=['ID','a0','b0','c0','d0','a1','b1','c1','d1','a2','b2','c2','d2','a3','b3','c3','d3','Average Daily Steps','No of active session','A_Mean','A_Median','A_Mode','No of sedentary session','S_Mean','S_Median','S_Mode'])



x = np.arange(24)
for index in d_c3.index:
    df = pd.read_csv('Dataset/FitnessDetails/Steps/'+str(d_c3.at[index,'ID'])+'.csv', index_col=0)
    sums = [0,0,0,0,0,0,0]
    mond= []
    tues= []
    wedn= []
    thur= []
    frid= []
    satu= []
    sund= []
    days = []
    A = []
    S = []
    for col in df.columns:  
        sums[0]+=df.at["Monday",col].mean()
        sums[1]+=df.at["Tuesday",col].mean()
        sums[2]+=df.at["Wednesday",col].mean()
        sums[3]+=df.at["Thursday",col].mean()
        sums[4]+=df.at["Friday",col].mean()
        sums[5]+=df.at["Saturday",col].mean()
        sums[6]+=df.at["Sunday",col].mean()
        mond.append(sums[0])
        tues.append(sums[1])
        wedn.append(sums[2])
        thur.append(sums[3])
        frid.append(sums[4])
        satu.append(sums[5])
        sund.append(sums[6])
    days.append(mond) 
    days.append(tues)
    days.append(wedn)
    days.append(thur)
    days.append(frid)
    days.append(satu)
    days.append(sund)
    for i in range(7):
        A = []
        S= []
        row = [d_c3.at[index,'ID']]
        nsedent=0
        nactive = 0
        y = days[i][:24]
        for j in range(0,24):
            if(y[j]-y[j-1]>400):
                nactive+=1
                A.append(y[j])
            elif(y[j]-y[j-1]<75):
                nsedent+=1
                S.append(y[j])
                
        cs = CubicSpline(x, y)
        for j in range(4):
            for k in range(3,-1,-1):
                row.append(cs.c.item(k,j))
        row.append(days[i][-1])
        row.append(nactive)
        row.append(np.mean(A))
        row.append(np.median(A))
        row.append(stats.mode(A)[0])
        row.append(nsedent)
        row.append(np.mean(S))
        row.append(np.median(S))
        row.append(stats.mode(S)[0])
        if i==0:
            mon_len = len(mon)
            mon.loc[mon_len] = row
        elif i==1:
            tue_len = len(tue)
            tue.loc[tue_len] = row
        elif i==2:
            wed_len = len(wed)
            wed.loc[wed_len] = row
        elif i==3:
            thu_len = len(thu)
            thu.loc[thu_len] = row
        elif i==4:
            fri_len = len(fri)
            fri.loc[fri_len] = row
        elif i==5:
            sat_len = len(sat)
            sat.loc[sat_len] = row
        elif i==6:
            sun_len = len(sun)
            sun.loc[sun_len] = row
        
mon.to_csv("Dataset/SplineFitted/Cluster 3/monday.csv",index=0)
tue.to_csv("Dataset/SplineFitted/Cluster 3/tuesday.csv",index=0)
wed.to_csv("Dataset/SplineFitted/Cluster 3/wednesday.csv",index=0)
thu.to_csv("Dataset/SplineFitted/Cluster 3/thursday.csv",index=0)
fri.to_csv("Dataset/SplineFitted/Cluster 3/friday.csv",index=0)
sat.to_csv("Dataset/SplineFitted/Cluster 3/saturday.csv",index=0)
sun.to_csv("Dataset/SplineFitted/Cluster 3/sunday.csv",index=0)
```
