```python
import pandas as pd


df = pd.read_csv("Dataset/physiological_features.csv", index_col = 0)
```

```python
import os
from math import sqrt

workout_par_rate = []
avg_wk_time=[]
std_wk_time=[]
avg_cal_burn = []
std_cal_burn = []
num_sessions = []

for id in os.listdir('Dataset/FitnessDetails/Calories'):
    df2 = pd.read_csv('Dataset/FitnessDetails/Calories/'+id)
    wpr = 0
    awk = 0
    swk = 0
    acb = 0
    scb = 0
    num = 0
    
    for index in df2.index:
        prev = 0
        for col in df2.columns:
            if col == "Total Calories" or col == "Day":
                continue
            
            if df2.at[index, col]>=150 :
                if prev == 0:
                    num+=1
                awk+= 1
                acb+= df2.at[index, col]
                prev = 1
            else:
                prev = 0
    if num!=0:
        acb = acb/num
        awk = awk/num
                
        for index in df2.index:
            prev = 0
            for col in df2.columns:
                if col == "Total Calories" or col == "Day":
                    continue

                if df2.at[index, col]>=150 :
                    swk+= abs(1-awk)*abs(1-awk)
                    scb+= abs(df2.at[index, col]-acb)*abs(df2.at[index, col]-acb)   
        swk = sqrt(swk/num)
        scb = sqrt(scb/num)
        wpr = num/30
    
    workout_par_rate.append(wpr) 
    avg_wk_time.append(awk)
    std_wk_time.append(swk) 
    avg_cal_burn.append(acb) 
    std_cal_burn.append(scb)
    num_sessions.append(num) 

    
df["Workout Par Rate"] = workout_par_rate
df["Average Workout Time"] = avg_wk_time
df["STD Workout Time"] = std_wk_time
df["Average Calories Burnt"] = avg_cal_burn
df["STD Calories Burnt"] = std_cal_burn
df["Number of Sessions"] = num_sessions
    
```

```python
df.to_csv("Dataset/contextual_features.csv")
```
