```python
import pandas as pd
import random
from random import randint

```

```python
import os
column = ['00-01 hr','01-02 hr','02-03 hr','03-04 hr','22-23 hr','23-24 hr']
cnt = 0
days = []
for i in range(30):
    
    if(cnt==0):
        days.append("Tuesday")
    elif(cnt==1):
        days.append("Wednesday")
    elif(cnt==2):
        days.append("Thursday")
    elif(cnt==3):
        days.append("Friday")
    else:
        days.append("Monday")
    cnt = (cnt+1)%5

for id in os.listdir('Temp'):
    if(id=="1116.csv" or id=="1117.csv" or id=="1118.csv"):
        continue
    df = pd.read_csv('Temp/'+id)
    for col in column:
        df[col] = df[col].apply(lambda x:randint(0,5))
    df = df.drop("Date",1)
    df.insert(loc=0, column='Day', value=days)
    df.to_csv("Temp/"+id,index = 0)
    
```

```python
import os
for id in os.listdir('Temp'):
    df = pd.read_csv('Temp/'+id,index_col=0)
    for col in df.columns:
        df[col] = df[col].apply(lambda x: int(x * 0.04))
    df.rename(columns = {'Total Steps':'Total Calories'}, inplace = True)
    df.to_csv('Calories/'+ id)
```
